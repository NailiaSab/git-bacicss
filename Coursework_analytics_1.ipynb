```python
import numpy as np #импортирование библиотек numpy, pandas,matplotlib
import pandas as pd
from datetime import datetime 
from datetime import date
import re
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')

def calculate_day(start_day): # функция для расчета даты рождения
    if '.' in start_day:
        start_day = datetime.strptime(start_day, "%Y-%m-%d %H:%M:%S.%f+00").date() 
    else:
        start_day = datetime.strptime(start_day, "%Y-%m-%d %H:%M:%S+00").date() 
    if start_day.month < 10:
        str_y_d = str(start_day.year)+'-'+'0'+str(start_day.month)
    else:
        str_y_d = str(start_day.year)+'-'+str(start_day.month)
    return str_y_d
        
courses = pd.read_csv('courses.csv') # перенос изначальных файлов в формате csv в DataFrame
course_contents = pd.read_csv('course_contents.csv')
progresses = pd.read_csv('progresses.csv')
progress_phases = pd.read_csv('progress_phases.csv')
progresses.rename(columns={'id': 'progress_id'}, inplace=True) # изменение название столбца для merge двух датасетов по столбцу
df1 =  progress_phases.merge(progresses, on = 'progress_id', how = 'left') #merge датасетов: progress_phases и progress
courses.rename(columns={'id': 'course_id'}, inplace=True)  # изменение название столбца для merge двух датасетов по столбцу
df2 = df1.merge(courses, on = 'course_id') #merge датасетов: df1 и courses
df2.rename(columns={'title': 'course_title'}, inplace=True) #изменение названия столбцов, как в примере
df2.rename(columns={'field': 'course_field'}, inplace=True)
df3 = df2.drop(columns = 'Unnamed: 0', axis='columns') #удаление ненужных столбцов
main = df3.merge(course_contents, on = ['lesson_number','module_number','course_id']) # merge двух датасетов по трем столбцам


df = main[(main['module_number'] == 1) & (main['is_homework'] == True)][['start_date','student_id','course_title']]
d1 = df.sort_values('start_date')


d1['start_course'] = d1['start_date'].apply(calculate_day)


kk = list(d1.groupby(['course_title','start_course']))

data_stud_rise_course = []
data_stud_rise_data = []
data_stud_rise_pers = []
for i in range(len(kk)):
    data_stud_rise_course.append(kk[i][0][0])
    data_stud_rise_data.append(kk[i][0][1])
    data_stud_rise_pers.append(len(kk[i][1]['student_id']))
data_result = {'Курс':data_stud_rise_course,'Дата':data_stud_rise_data, 'Количество':data_stud_rise_pers}
df = pd.DataFrame(data_result) #график с количеством прироста в каждый месяц человек на курсе
df.loc[len(df.index)] = ['Веб-дизайн Базовый','2016-03',0]
df.loc[len(df.index)] = ['Веб-дизайн Базовый','2019-07',0]
df_graph = df.sort_values('Дата')
dff = df_graph.sort_index()
print(dff)
group_courses = dff.groupby('Курс')
result = list(group_courses)
print(len(result))
i = 0       
fig, axes = plt.subplots(figsize=(10,150),ncols= 1, nrows = len(result))
for ax in axes:
    sns.lineplot(data=result[i][1], x='Дата', y='Количество', ax = ax)
    ax.set_title('Курс: {}'.format(result[i][0]))
    ax.set_xlabel('Время')
    ax.set_ylabel('Количество учеников')
    ax.tick_params(axis = 'x', rotation = 90)
    i += 1
fig.suptitle('Динамика прироста студентов в каждом месяце для каждого курса',y = 0.89, fontsize=20 )



fig, ax = plt.subplots(figsize = (20,20)) #добавление плота
ax = sns.lineplot(data=df_graph, x='Дата', y = 'Количество', hue='Курс') #создание диаграммы
ax.set_title('Динамика прироста студентов в каждом месяце для каждого курса', size = 15) #добавление названия графика
ax.set_ylabel('Количество студентов') #добавление подписи по оси x
ax.set_xlabel('Время')#добавление подписи по оси y
ax.tick_params(axis = 'x', rotation = 90)
plt.show()


df = main[main['is_homework'] == True][['start_date','progress_id','course_title']]
d1 = df.sort_values('start_date')


d1['start_course'] = d1['start_date'].apply(calculate_day)


kk = list(d1.groupby(['course_title','start_course']))

data_stud_rise_course = []
data_stud_rise_data = []
data_stud_rise_pers = []
for i in range(len(kk)):
    data_stud_rise_course.append(kk[i][0][0])
    data_stud_rise_data.append(kk[i][0][1])
    data_stud_rise_pers.append(len(kk[i][1]['progress_id']))
data_result = {'Курс':data_stud_rise_course,'Дата':data_stud_rise_data, 'Количество прогрессов':data_stud_rise_pers}
df = pd.DataFrame(data_result) #график с количеством прироста в каждый месяц человек на курсе
df_graph = df.sort_values('Дата')
dff = df_graph.sort_index()
print(dff)


group_courses = dff.groupby('Курс')
result = list(group_courses)
print(len(result))

i = 0       
fig, axes = plt.subplots(figsize=(10,150),ncols= 1, nrows = len(result))
for ax in axes:
    sns.lineplot(data=result[i][1], x='Дата', y='Количество прогрессов', ax = ax)
    ax.set_title('Курс: {}'.format(result[i][0]))
    ax.set_xlabel('Время')
    ax.set_ylabel('Количество прогрессов')
    ax.tick_params(axis = 'x', rotation = 90)
    i += 1
fig.suptitle('Динамика прогрессов в каждом месяце для каждого курса',y = 0.89, fontsize=20 )



fig, ax = plt.subplots(figsize = (20,20)) #добавление плота
ax = sns.lineplot(data=df_graph, x='Дата', y = 'Количество прогрессов', hue='Курс') #создание диаграммы
ax.set_title('Динамика прогрессов в каждом месяце для каждого курса', size = 15) #добавление названия графика
ax.set_ylabel('Количество прогрессов') #добавление подписи по оси x
ax.set_xlabel('Время')#добавление подписи по оси y
ax.tick_params(axis = 'x', rotation = 90)
plt.show()
```
