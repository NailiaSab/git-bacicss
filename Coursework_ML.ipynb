# 1. Описание данных

```python
import pandas as pd # импортирование библиотеки pandas

dataset = pd.read_csv('telecom_users.csv')  # прочтение файл формата csv и создание датасета
dataset.head(10) #вывод первых 20 строк датасета
```

```python
# Приведение названий столбцов к единому формату имен с заглавной буквы
dataset.rename(columns={'customerID':'CustomerID','gender': 'Gender', 'tenure': 'Tenure'}, inplace=True)
```

```python
# В столбце TotalCharges тип данных object, нужно поменять тип данных на float
# Для того, чтобы не трогать основной датасет, скопируем его 
dataset_1 = dataset.copy()
# Приведение числовых данных к правильному типу
new_float = []
for num in dataset_1['TotalCharges']:
    try:
      num = float(num)
    except:
      num = 0
    new_float.append(num)
dataset_1['TotalCharges'] = new_float
```

```python
# Проверка уникальных значений в столбце CustomerID
dataset_1['CustomerID'].value_counts()
```

```python
# Все значения уникальные в столбце CustomerID, так как эта переменная не несет никакой нагрузки ее можно удалить 
dataset_1.drop(columns={'CustomerID'}, inplace=True)
dataset_1.info()
```

# 3. Построение моделей для прогнозирования оттока

```python
# Разобьем наши данные на тренировочную выборку и тестовую
from sklearn.model_selection import train_test_split
# Определение тренировочной и тестовую выборку
x_train, x_test, y_train, y_test = train_test_split(
    new_df.values, df2.iloc[:, -1].values, test_size=0.3,
    random_state=17)
```

```python
# Нормализация данных для лучшего обучения алгоритмов
from sklearn.preprocessing import  MinMaxScaler
x_train = MinMaxScaler().fit_transform(x_train)
x_test = MinMaxScaler().fit_transform(x_test)
```

```python
# импортирование алгоритмов
from sklearn.model_selection import (GridSearchCV,
                                     StratifiedKFold)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression




# параметры базовых алгоритмов
gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)} # GradientBoostingClassifier

rfc_params = {'n_estimators': range(10, 100, 10), # RandomForestClassifier
              'min_samples_leaf': range(1, 5)}

svc_params = {'kernel': ['linear', 'rbf'], # SVC
              'C': np.arange(0.1, 1, 0.2)}

lr_params = {'C': np.arange(0.5, 1, 0.1)}  # LR

knn_params = {'n_neighbors' : np.arange(1, 10, 1)} # KNN

skf = StratifiedKFold(n_splits=5, random_state=17, shuffle = True) # k-фолдовая кратность
```

```python
#  Определение объекта GridSearchCV для всех приведенных параметров каждого алгоритма
svc = GridSearchCV(SVC(random_state=17), svc_params, cv = skf)
lr = GridSearchCV(LogisticRegression(random_state=17), lr_params, cv = skf)
gbc =GridSearchCV(GradientBoostingClassifier( random_state=17), gbc_params, cv = skf)
rfc = GridSearchCV(RandomForestClassifier( random_state=17), rfc_params, cv = skf)
knn = GridSearchCV(KNeighborsClassifier(), knn_params, cv = skf)

# создание листа из алгоритмо
algorithms = [svc,lr,gbc,rfc,knn] 

# Оптимизация гиперпараметров с помощью поиска по сетке
# Обучение на тренировочной выборке и выведение лучших параметров
for i in algorithms:
  i.fit(x_train, y_train)
  print('Лучшие сочетание параметров для алгоритма ',i,'\n\n',  i.best_params_, '\n')
```

```python
# Метрика accuracy будет неактульна для сравнения качетсва моделей, так как бесполезна в задачах с неравными классами
# Поэтому ее как метрику для сравнения качетсва моделей использовать не будем, так как 
# Процент оттока клиентов составляет примерно: 26,55%, что говорит о разном количестве  Yes и No в таргете Churn

# Импортируем метрики
from sklearn.metrics import (recall_score,
                             precision_score,
                             f1_score,
                             roc_auc_score)


# Получение значений полноты моделей классификаторов
metrics_scores = [recall_score, precision_score,f1_score]
predicts = [ gbc_predict, rfc_predict, svc_predict, lr_predict, knn_predict]
models_names = ['GradientBoosting', 'RandomForest', 'SVC', 'LogisticRegression', 'KNeighbors']
scores_names = ['recall_score', 'precision_score','f1_score']
values_list = []
for i, score in enumerate(metrics_scores):
    for predict in predicts:
        values_list.append(round(score(y_test, predict),3))    
    
    x = np.arange(len(models_names))  # Список координат столбцов по x
    
    fig, ax = plt.subplots(figsize=(10,5))  # Определение фигуры и осей
    rects = ax.bar(x, values_list, 0.6)  # Определение колонок для данных
    
    # Указание подписей для осей, таблицы, легенды
    ax.set_ylabel('Значение метрики')
    ax.set_xlabel('Модели')
    ax.set_title(f'Значения {scores_names[i]} для моделей классификации')
    ax.set_xticks(x)
    ax.set_xticklabels(models_names)
    
    # Определение функции для отображения столбцов с аннотациями
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            # Задание параметров для аннотаций
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height), # Получение точек координат для текста
                        xytext=(0, 2),    # Высота текста над столбцами
                        # Расположение текста относительно столбцов
                        textcoords="offset points",  
                        ha='center', va='bottom')
    
    autolabel(rects)  # Выполнение функции
    values_list = []
# вывод плота
plt.show()
```
